```python
**Take employee name
Take salary input (with validation)
Calculate:
HRA = 20%
Tax:

50k → 10%

30k → 5%

else → 2%
Calculate net salary
Show full summary
Handle all errors
**
```

```python
try:
    Emp_name=input("enter the empployee name:")
    Sal=float(input("enter the total salary:"))
    HRA=(Sal*20)/100
    if Sal>=50000:
        tax=(Sal*10)/100
    elif Sal>=30000:
        tax=(Sal*5)/100
    else:
        tax=(Sal*2)/100
    #net_sal
    net_sal=Sal+HRA-tax
    print("\n----- Salary Summary -----")
    print("Employee Name:", Emp_name)
    print("Basic Salary:", Sal)
    print("HRA:", HRA)
    print("Tax:", tax)
    print("Net Salary:", net_sal)
except ValueError:
    print("Invalid salary input! Please enter numeric value.")
    
```

**Take student name
Take marks for 3 subjects
Validate marks (0–100)
Calculate:
Total
Average
Grade
Handle invalid input
Print result summary**

```python
try:
    std_name=input("Enter the name of std:")
    sub_1=float(input("enter the number of Sub_1:"))
    sub_2=float(input("enter the number of Sub_2:"))
    sub_3=float(input("enter the number of Sub_3:"))
#validation
    if (sub_1<0 or sub_1>100 or
        sub_2<0 or sub_2>100 or
        sub_3<0 or sub_3>100):
         print("Invalid marks! Marks should be between 0 and 100.")
    else:
        total=sub_1+sub_2+sub_3
        avg=total/3
        #grade_logic
        if avg>90:
            grade="A"
        elif avg>=75:
            grade="B"
        elif avg>=60:
            grade="C"
        else:
            grade="fail"
        # output
        print("\n----- Result Summary -----")
        print("Student Name:", std_name)
        print("Total Marks:", total)
        print("Average:", avg)
        print("Grade:", grade)
except ValueError:
    print("Invalid input! Please enter numeric marks.")
```
