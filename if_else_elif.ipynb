```python
#write program enter a percentage
#% > 75 print A grade
#if % > between 60 and 70 print B grade
#if % > between 45 and 60 print C grade
#if % > between 35 and 45 print D grade
#else Fail
percentage=eval(input("Enter the percentage"))
if percentage>=75:
    print("A grade")
elif percentage>=60 and percentage<=70:
    print("B grade")
elif percentage>=45 and percentage<=60:
    print("C grade")
elif percentage>=35 and percentage<=45:
    print("D grade")
else:
    print("Fail")
```
