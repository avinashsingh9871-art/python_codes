```python
#ask user to enter a number
#if the number >=o
#then check if that number >=0 then print zero
#else print positive
# otherwise print negative
num=eval(input("enter a number:"))
if num>=0:
    if num==0:
        print("its a Zero")
    else:
        print(f"the number is {num} Postive")
else:
    print(f"the number is {num} negative")
```

```python
#ask you to print a number if it is even or odd
#even number % by 2, the reminder will be 0
#odd number % by 2, the reminder will not be 0
#/ // % math operators
num=eval(input("enter a number:"))
if num % 2 == 0:
    print(f"the {num} entered by the user is even")
else:
    print(f"the {num} entered by the user is Odd")
```

```python
#gov already providing 25 km for free, and you need to travel for 100 km, then we need to make 75 km fare
import time
user_dis=eval(input("Enter how much KM you nee dto travel:"))
govt_free_dis=25
if user_dis>govt_free_dis:
    per_KM_charges=eval(input("ente rthe amount per KM"))
    pending_dis=user_dis-govt_free_dis
    pending_amt_pay=per_KM_charges*pending_dis
    time.sleep(2)
    print(f"total charges user need pay {pending_amt_pay}")
else:
    print("enjoy your free ride")
    
```

```python
#ask the user to get a random integer between 1 and 10
#assign as a number n1
#ask user to enter the number: n2
#if both are matched, print " You win
#else you lost
import random
User_n1=random.randint(1,10)
User_n2=eval(input("enter number between 1 to 10"))
if User_n1==User_n2:
    print(f"user 1st number was {User_n1}")
    print(f"user 2nd number was {User_n2}")
    print(f"you have won the game")
else:
    print(f"user 1st number was {User_n1}")
    print(f"user 2nd number was {User_n2}")
    print("you have lost the game")

```
